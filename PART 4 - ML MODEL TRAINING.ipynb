# PREPARING DATA FOR TRAINING THE MODEL 

```python
x=data.drop('Result',axis=1)
y=data['Result']
```

```python
#DIVIDING THE DATA FOR TRAINING AND TESTING THE MODEL
X_train,X_test,Y_train,Y_test = train_test_split(x,y,train_size=0.75,random_state=31)
```

```python
print("The Shape of Xtrain is ",X_train.shape)
print("The Shape of Xtest is ",X_test.shape)
print("The Shape of Ytrain is ",Y_train.shape)
print("The Shape of Ytest is ",Y_test.shape)
```

```python
#SCALING THE VALUES IN A PARTICULAR RANGE : STANDARDISATION
SS=StandardScaler()
X_train=SS.fit_transform(X_train)
X_test=SS.fit_transform(X_test)
```

# MODEL TRAINING BY LOGISTIC REGRESSION

```python

# Necessary imports
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
 
# Creating the hyperparameter grid
c_space = np.logspace(-5, 8, 15)
param_grid = {'C': c_space}
 
# Instantiating logistic regression classifier
logreg = LogisticRegression()
 
# Instantiating the GridSearchCV object
logreg_cv = GridSearchCV(logreg, param_grid, cv = 5)
 
logreg_cv.fit(X_train, Y_train)
 
# Print the tuned parameters and score
print("Tuned Logistic Regression Parameters: {}".format(logreg_cv.best_params_))
print("Best score is {}".format(logreg_cv.best_score_))
```

```python
#TRAINING THE MODEL 
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,recall_score,precision_score,r2_score
from sklearn.ensemble import RandomForestClassifier
Logreg = LogisticRegression()
Logreg.fit(X_train,Y_train)
Y_pred = Logreg.predict(X_test)
RFC=accuracy_score(Y_test,Y_pred)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test,Y_pred)
TP = cm[0,0]
TN = cm[0,1]
FP = cm[1,0]
FN = cm[1,1]
print("LOGISTIC REGRESSION")
print('TP:',TP ,'\nTN:',TN,'\nFP:',FP,'\nFN:', FN)
print('Presicion Score is: ',precision_score(Y_test,Y_pred))
print('Recall score is:',recall_score(Y_test,Y_pred))
print('Accuaracy Score is:',accuracy_score(Y_test,Y_pred))
```
