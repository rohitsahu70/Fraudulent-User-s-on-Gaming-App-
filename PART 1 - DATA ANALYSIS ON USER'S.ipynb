```python
'''
THIS FUNCTION IS USED FOR REMOVE ALL THE DATA WHICH CONSIST OF 0'S VALUES IN DAYS SPENT, DEPOSITE COUNT, 
DEPOSITE AMOUNT AND REDEEM COUNT.
'''
def preprocessing(df):
    df=df[df.DaysSpent>0]
    df = df[df.DepositCount>0]
    df = df[df.DepositAmount>0]
    df = df[df.RedeemCount>0]
    return df
```

# BLOCKED DATA

```python
def meanof5to50days(df):
    days=[5,10,15,20,25,30,35,40,45,50]
    list1=[]
    list2=[]
    list3=[]
    list4=[]
    list5=[]
    dict={0:5,6:10,11:15,16:20,21:25,26:30,31:35,36:40,41:45,46:50}
    for i,val in dict.items():
        temp=df[(df.DaysSpent>=i) & (df.DaysSpent<=val)]
        val1=temp.DepositCount.mean()
        list1.append(val1)
        val2=temp.DepositAmount.mean()
        list2.append(val2)
        val3=temp.RedeemCount.mean()
        list3.append(val3)
        val4=temp.RedeemAmount.mean()
        list4.append(val4)
        val5=temp.PaidGamePlayedWinRatio.mean()
        list5.append(val5)
        temp=df
    data1=pd.DataFrame(list(zip(list1,list2,list3,list4,list5)),columns=["DepositCount","DepositAmount","RedeemCount","RedeemAmount","winratio"],index=days)
    return data1
block_data=meanof5to50days(df1)
block_data
```

# BLOCKED DATA WITH PREPROCESSING

```python
temp=preprocessing(df1)
preprocess_block_data=meanof5to50days(temp)
preprocess_block_data
```

```python
'''
THIS FUNCTION RETURNS A DATA FRAME WHICH SHOWS HOW MANY TIMES A MEAN OF BLOCK USER'S DATA IS GREATER THAN THE MEAN
OF NORMAL USER'S DATA.
'''
def meanofdata(df,df1,start,end,interval):
    days=[]
    for i in range(start,end+interval,interval):
        days.append(i)
    DAYS=[]
    for i in range(len(days)-1):
        DAYS.append("{} - {}".format(days[i],days[i+1]))

#   UNBLOCK USER
    list1,list2,list3,list4,list5=[],[],[],[],[]
    for i in range(len(days)-1):
        temp=df[(df.DaysSpent>=days[i]+1) & (df.DaysSpent<=days[i+1])]
        val1=temp.DepositCount.mean()
        list1.append(val1)
        val2=temp.DepositAmount.mean()
        list2.append(val2)
        val3=temp.RedeemCount.mean()
        list3.append(val3)
        val4=temp.RedeemAmount.mean()
        list4.append(val4)
        val5=temp.PaidGamePlayedWinRatio.mean()
        list5.append(val5)
        temp=df
    dataframe1=pd.DataFrame(list(zip(DAYS,list1,list2,list3,list4,list5)),columns=["Days","meanofDepositCount","meanofDepositAmount","meanofRedeemCount","meanofRedeemAmount","meanofwinratio"])

#   BLOCK USER
    list1,list2,list3,list4,list5=[],[],[],[],[]
    for i in range(len(days)-1):
        temp=df1[(df1.DaysSpent>=days[i]+1) & (df1.DaysSpent<=days[i+1])]
        val1=temp.DepositCount.mean()
        list1.append(val1)
        val2=temp.DepositAmount.mean()
        list2.append(val2)
        val3=temp.RedeemCount.mean()
        list3.append(val3)
        val4=temp.RedeemAmount.mean()
        list4.append(val4)
        val5=temp.PaidGamePlayedWinRatio.mean()
        list5.append(val5)
        temp=df
    dataframe2=pd.DataFrame(list(zip(DAYS,list1,list2,list3,list4,list5)),columns=["Days","meanofDepositCount","meanofDepositAmount","meanofRedeemCount","meanofRedeemAmount","meanofwinratio"])

#   COMPARISION BETWEEN BLOCK AND UNBLOCKED DATA
    list1=dataframe2.meanofDepositCount/dataframe1.meanofDepositCount
    list2=dataframe2.meanofDepositAmount/dataframe1.meanofDepositAmount
    list3=dataframe2.meanofRedeemCount/dataframe1.meanofRedeemCount
    list4=dataframe2.meanofRedeemAmount/dataframe1.meanofRedeemAmount
    list5=dataframe2.meanofwinratio/dataframe1.meanofwinratio
    finaldf=pd.DataFrame(list(zip(DAYS,list1,list2,list3,list4,list5)),columns=["days","DepositCount","DepositAmount","RedeemCount","RedeemAmount","winratio"])
    return finaldf
```
