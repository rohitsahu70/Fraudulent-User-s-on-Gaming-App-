# DATA PREPROCESSING 

```python
def preprocessing(df):
    df = df[df.AppLaunchCount>0]
    df = df[df.DaysSpent>0]
    df = df[df.DepositCount>0]
    df = df[df.DepositAmount>0]
    df = df[df.RedeemCount>0]
    return df
```

# COMOPARING FUNCTION

```python
def meanofdata(df,df1,start,end,interval):
    
    days=[]
    for i in range(start,end+interval,interval):
        days.append(i)
    
    DAYS=[]
    for i in range(len(days)-1):
        DAYS.append("{} - {}".format(days[i],days[i+1]))

#   UNBLOCK USER
    list1,list2,list3,list4,list5=[],[],[],[],[]
    for i in range(len(days)-1):
        temp=df[(df.DaysSpent>=days[i]+1) & (df.DaysSpent<=days[i+1])]
        val1=temp.DepositCount.mean()
        list1.append(val1)
        val2=temp.DepositAmount.mean()
        list2.append(val2)
        val3=temp.RedeemCount.mean()
        list3.append(val3)
        val4=temp.RedeemAmount.mean()
        list4.append(val4)
        val5=temp.PaidGamePlayedWinRatio.mean()
        list5.append(val5)
        temp=df
    dataframe1=pd.DataFrame(list(zip(DAYS,list1,list2,list3,list4,list5)),columns=["Days","meanofDepositCount","meanofDepositAmount","meanofRedeemCount","meanofRedeemAmount","meanofwinratio"])

#   BLOCK USER
    list1,list2,list3,list4,list5=[],[],[],[],[]
    for i in range(len(days)-1):
        temp=df1[(df1.DaysSpent>=days[i]+1) & (df1.DaysSpent<=days[i+1])]
        val1=temp.DepositCount.mean()
        list1.append(val1)
        val2=temp.DepositAmount.mean()
        list2.append(val2)
        val3=temp.RedeemCount.mean()
        list3.append(val3)
        val4=temp.RedeemAmount.mean()
        list4.append(val4)
        val5=temp.PaidGamePlayedWinRatio.mean()
        list5.append(val5)
        temp=df
    dataframe2=pd.DataFrame(list(zip(DAYS,list1,list2,list3,list4,list5)),columns=["Days","DepositCount","DepositAmount","RedeemCount","RedeemAmount","winratio"])

#   COMPARISION BETWEEN BLOCK AND UNBLOCKED DATA
    list1=dataframe2.DepositCount/dataframe1.meanofDepositCount
    list2=dataframe2.DepositAmount/dataframe1.meanofDepositAmount
    list3=dataframe2.RedeemCount/dataframe1.meanofRedeemCount
    list4=dataframe2.RedeemAmount/dataframe1.meanofRedeemAmount
    list5=dataframe2.winratio/dataframe1.meanofwinratio
    finaldf=pd.DataFrame(list(zip(DAYS,list1,list2,list3,list4,list5)),columns=["days","DepositCount","DepositAmount","RedeemCount","RedeemAmount","winratio"])
    return finaldf
temp=preprocessing(df)
temp1=preprocessing(df1)
start=0
end=50
interval=5
finaldf=meanofdata(temp,temp1,start,end,interval)
```

# ANOMALY DETECTION

```python
def frauddetection(userid,k,DataFrame):
    count=0
    #blocked user 
#     user=df1[df1['USERIDs']==userid]
    #unblocked user
    user=df[df['USERIDs']==userid]
    
    days=user.DaysSpent.iloc[0]
    
    #FIRST
    if days >=0 or days<5:
        if user.DepositCount.iloc[0] > DataFrame.DepositCount.iloc[0]/k:
            count+=1
        if user.DepositAmount.iloc[0] > DataFrame.DepositAmount.iloc[0]/k:
            count+=1
        if user.RedeemCount.iloc[0] > DataFrame.RedeemCount[0]/k:
            count+=1
        if user.RedeemAmount.iloc[0] > DataFrame.RedeemAmount[0]/k:
            count+=1
        if user.PaidGamePlayedWinRatio.iloc[0]>DataFrame.winratio[0]/k:
            count+=1
    
    # SECOND
    elif days >=5 or days<10:
        if user.DepositCount.iloc[0] > DataFrame.DepositCount.iloc[1]/k:
            count+=1
        if user.DepositAmount.iloc[0] > DataFrame.DepositAmount.iloc[1]/k:
            count+=1
        if user.RedeemCount.iloc[0] > DataFrame.RedeemCount[1]/k:
            count+=1
        if user.RedeemAmount.iloc[0] > DataFrame.RedeemAmount[1]/k:
            count+=1
        if user.PaidGamePlayedWinRatio.iloc[0]>DataFrame.winratio[1]/k:
            count+=1
    
    # THIRD
    elif days >= 10 or days<15:
        if user.DepositCount.iloc[0] > DataFrame.DepositCount.iloc[2]/k:
            count+=1
        if user.DepositAmount.iloc[0] > DataFrame.DepositAmount.iloc[2]/k:
            count+=1
        if user.RedeemCount.iloc[0] > DataFrame.RedeemCount[2]/k:
            count+=1
        if user.RedeemAmount.iloc[0] > DataFrame.RedeemAmount[2]/k:
            count+=1
        if user.PaidGamePlayedWinRatio.iloc[0]>DataFrame.winratio[2]/k:
            count+=1
    
    # FOURTH
    elif days>=15 or days<20:
        if user.DepositCount.iloc[0] > DataFrame.DepositCount.iloc[3]/k:
            count+=1
        if user.DepositAmount.iloc[0] > DataFrame.DepositAmount.iloc[3]/k:
            count+=1
        if user.RedeemCount.iloc[0] > DataFrame.RedeemCount[3]/k:
            count+=1
        if user.RedeemAmount.iloc[0] > DataFrame.RedeemAmount[3]/k:
            count+=1
        if user.PaidGamePlayedWinRatio.iloc[0]>DataFrame.winratio[3]/k:
            count+=1

    #FIVE			
    elif days>=20 or days<25:
        if user.DepositCount.iloc[0] > DataFrame.DepositCount.iloc[4]/k:
            count+=1
        if user.DepositAmount.iloc[0] > DataFrame.DepositAmount.iloc[4]/k:
            count+=1
        if user.RedeemCount.iloc[0] > DataFrame.RedeemCount[4]/k:
            count+=1
        if user.RedeemAmount.iloc[0] > DataFrame.RedeemAmount[4]/k:
            count+=1
        if user.PaidGamePlayedWinRatio.iloc[0]>DataFrame.winratio[4]/k:
            count+=1
            
    #SIXTH			
    elif days>=25 or days<30:
        if user.DepositCount.iloc[0] > DataFrame.DepositCount.iloc[5]/k:
            count+=1
        if user.DepositAmount.iloc[0] > DataFrame.DepositAmount.iloc[5]/k:
            count+=1
        if user.RedeemCount.iloc[0] > DataFrame.RedeemCount[5]/k:
            count+=1
        if user.RedeemAmount.iloc[0] > DataFrame.RedeemAmount[5]/k:
            count+=1
        if user.PaidGamePlayedWinRatio.iloc[0]>DataFrame.winratio[5]/k:
            count+=1

    #SEVENTH 			
    elif days>=30 or days<35:
        if user.DepositCount.iloc[0] > DataFrame.DepositCount.iloc[6]/k:
            count+=1
        if user.DepositAmount.iloc[0] > DataFrame.DepositAmount.iloc[6]/k:
            count+=1
        if user.RedeemCount.iloc[0] > DataFrame.RedeemCount[6]/k:
            count+=1
        if user.RedeemAmount.iloc[0] > DataFrame.RedeemAmount[6]/k:
            count+=1
        if user.PaidGamePlayedWinRatio.iloc[0]>DataFrame.winratio[6]/k:
            count+=1

    #EIGHT 			
    elif days>=35 or days<40:
        if user.DepositCount.iloc[0] > DataFrame.DepositCount.iloc[7]/k:
            count+=1
        if user.DepositAmount.iloc[0] > DataFrame.DepositAmount.iloc[7]/k:
            count+=1
        if user.RedeemCount.iloc[0] > DataFrame.RedeemCount[7]/k:
            count+=1
        if user.RedeemAmount.iloc[0] > DataFrame.RedeemAmount[7]/k:
            count+=1
        if user.PaidGamePlayedWinRatio.iloc[0]>DataFrame.winratio[7]/k:
            count+=1
            
    if count>=4:
        return "the user is a fraud user"
    else:
        return "the user is not fraud"
```
